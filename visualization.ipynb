```python
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
distilbert_model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')
```

```python
neg_sents = ["This is awful!",
            "I genuinely hate this movie.",
            "These allegations are terrible.",
            "How could you be so stupid?"]
neg = distilbert_model.encode(neg_sents)

pos_sents = ["I'd love to be there for you!",
            "They were always so delighted in the mornings.",
            "She was constantly smiling during her stay.",
            "I'm pretty confident that this was the best day ever."]
pos = distilbert_model.encode(pos_sents)
```

```python
sents = [row for row in pos]
[sents.append(row) for row in neg]
sents = np.asarray(sents)
```

```python
# Reduce to 3 dimensions using PCA and t-SNE
pca = PCA(n_components=3)
pca_results = pca.fit_transform(sents)

tsne = TSNE(n_components=3)
tsne_results = tsne.fit_transform(sents)
```

```python
# Make a scatter-plot in two dimensions using the representations of PCA and t-SNE
fig, ax = plt.subplots()
ax.scatter(x=pca_results[:4,0], y=pca_results[:4,1],color='red',label='Negative')
ax.scatter(x=pca_results[4:,0], y=pca_results[4:,1],color='blue',label='Positive')
ax.legend()
ax.set_title("PCA of distilbert")
fig.show()

fig, ax = plt.subplots()
ax.scatter(x=tsne_results[:4,0], y=tsne_results[:4,1],color='red',label='Negative')
ax.scatter(x=tsne_results[4:,0], y=tsne_results[4:,1],color='blue',label='Positive')
ax.legend()
ax.set_title("t-SNE of distilbert")
fig.show()
```
